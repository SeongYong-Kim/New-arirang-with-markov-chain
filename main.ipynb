```python
myMIDI = MIDIFile(1)
myMIDI.addTrackName(0,0,"Sample Track")
myMIDI.addTempo(0,0,120)

c_time = 0
for note in SeongbukArirang:
    if note[0] == 'r':
        myMIDI.addNote(0, 0, pitch=0, time=c_time, duration=float(note[1]), volume=0)
    else:
        myMIDI.addNote(0, 0, int(note[0]), c_time, float(note[1]), 100)

    c_time  = c_time + float(note[1])

#(트랙, 채널, pitch, time, duration, volume)
binfile = open("SeongbukArirang.mid", 'wb')
myMIDI.writeFile(binfile)
binfile.close()
```
